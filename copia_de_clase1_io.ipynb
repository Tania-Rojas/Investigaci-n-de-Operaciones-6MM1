Dibujo de redes

```python
#04/09
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G=nx.DiGraph()
G.add_edge('A','B', capacity=2)
G.add_edge('A','C', capacity=2)
G.add_edge('B','D', capacity=2)
G.add_edge('C','D', capacity=2)
G.add_edge('B','C', capacity=2)
nx.draw_networkx(G) #Dibuja
plt.title('RED:Arcos y nodos')
plt.figure()
nx.nodes(G) #nos da los nodos
nx.edges(G) #nos da los arcos
```
